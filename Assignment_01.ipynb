# *01_double_it*

```python
def main():
    # Ask the user to enter a number and convert it to an integer
    curr_value = int(input("Enter a number: "))

    # Double the number repeatedly until it's 100 or more
    while curr_value < 100:
        curr_value = curr_value * 2
        print(curr_value)

# This provided line is required at the end of
# Python file to call the main() function.
if __name__ == '__main__':
    main()
```

# *03_guess_my_number*

```python
import random

def main():
    # Generate the secret number at random!
    secret_number: int = random.randint(1, 99)

    print("I am thinking of a number between 1 and 99...")

    # Get user's guess
    guess = int(input("Enter a guess: "))
    # True if guess is not equal to secret number
    while guess != secret_number:
        if guess < secret_number:  # If-statement is True if guess is less than secret number
            print("Your guess is too low")
        else:
            print("Your guess is too high")

        print() # Print an empty line to tidy up the console for new guesses
        guess: int = int(input("Enter a new guess: "))  # Get a new guess from the user

    print("Congrats! The number was: " + str(secret_number))

if __name__ == '__main__':
    main()
```

## **Control Flow for Console**

# *Solution*

```python
import random

NUM_ROUNDS = 5
MIN_VALUE = 1
MAX_VALUE = 100

def get_user_guess():
    guess = input("Do you think your number is higher or lower than the computer's?: ").strip().lower()
    while guess not in ['higher', 'lower']:
        guess = input("Please enter either 'higher' or 'lower': ").strip().lower()
    return guess

def main():
    print("Welcome to the High-Low Game!")
    print("--------------------------------")

    score = 0

    for round_num in range(1, NUM_ROUNDS + 1):
        print(f"Round {round_num}")
        user_num = random.randint(MIN_VALUE, MAX_VALUE)
        computer_num = random.randint(MIN_VALUE, MAX_VALUE)

        print(f"Your number is {user_num}")
        guess = get_user_guess()

        if user_num == computer_num:
            print(f"Aww, that's incorrect. The computer's number was {computer_num}")
        elif guess == 'higher' and user_num > computer_num:
            print(f"You were right! The computer's number was {computer_num}")
            score += 1
        elif guess == 'lower' and user_num < computer_num:
            print(f"You were right! The computer's number was {computer_num}")
            score += 1
        else:
            print(f"Aww, that's incorrect. The computer's number was {computer_num}")

        print(f"Your score is now {score}\n")

    print("Thanks for playing!")

    # Extension #2: Conditional ending messages
    if score == NUM_ROUNDS:
        print("Wow! You played perfectly!")
    elif score >= NUM_ROUNDS // 2:
        print("Good job, you played really well!")
    else:
        print("Better luck next time!")

if __name__ == '__main__':
    main()
```

## *marsweight*

```python
"""
Prompts the user for a weight on Earth
and prints the equivalent weight on Mars.
"""

# Constant for Mars gravity relative to Earth
MARS_MULTIPLE = 0.378

def main():
    # Step 1: Ask the user for their Earth weight
    earth_weight_str = input("Enter a weight on Earth: ")

    # Step 2: Convert string to float for calculation
    earth_weight = float(earth_weight_str)

    # Step 3: Calculate weight on Mars
    mars_weight = earth_weight * MARS_MULTIPLE

    # Step 4: Round result to 2 decimal places
    rounded_mars_weight = round(mars_weight, 2)

    # Step 5: Print the final result
    print("The equivalent weight on Mars: " + str(rounded_mars_weight))

# Run the program
if __name__ == "__main__":
    main()
```

## *planetaryweight*

```python
"""
Prompts the user for a weight on Earth
and a planet (in separate inputs). Then
prints the equivalent weight on that planet.

Assumes correct capitalization and a valid planet name.
"""

# Constants for each planet's gravity relative to Earth
MERCURY_GRAVITY = 0.376
VENUS_GRAVITY = 0.889
MARS_GRAVITY = 0.378
JUPITER_GRAVITY = 2.36
SATURN_GRAVITY = 1.081
URANUS_GRAVITY = 0.815
NEPTUNE_GRAVITY = 1.14

def main():
    # Step 1: Get user input for Earth weight
    earth_weight = float(input("Enter a weight on Earth: "))

    # Step 2: Get user input for planet
    planet = input("Enter a planet: ")

    # Step 3: Determine gravity multiplier using if-elif-else
    if planet == "Mercury":
        gravity_constant = MERCURY_GRAVITY
    elif planet == "Venus":
        gravity_constant = VENUS_GRAVITY
    elif planet == "Mars":
        gravity_constant = MARS_GRAVITY
    elif planet == "Jupiter":
        gravity_constant = JUPITER_GRAVITY
    elif planet == "Saturn":
        gravity_constant = SATURN_GRAVITY
    elif planet == "Uranus":
        gravity_constant = URANUS_GRAVITY
    else:
        # Neptune assumed for any other valid planet input
        gravity_constant = NEPTUNE_GRAVITY

    # Step 4: Calculate planetary weight
    planetary_weight = earth_weight * gravity_constant

    # Step 5: Round result to 2 decimal places
    rounded_weight = round(planetary_weight, 2)

    # Step 6: Output the final result
    print("The equivalent weight on " + planet + ": " + str(rounded_weight))

# Run the program
if __name__ == "__main__":
    main()
```

# *index_game*

```python
"""
A small interactive program that allows users to access,
modify, or slice a predefined list using list indexing.
"""

def access_element(lst, index):
    try:
        return lst[index]
    except IndexError:
        return "Index out of range."

def modify_element(lst, index, new_value):
    try:
        lst[index] = new_value
        return lst
    except IndexError:
        return "Index out of range."

def slice_list(lst, start, end):
    try:
        return lst[start:end]
    except IndexError:
        return "Invalid indices."

def index_game():
    # Step 1: Define a sample list
    lst = [1, 2, 3, 4, 5]
    print("Current list:", lst)

    # Step 2: Ask user for operation
    print("Choose an operation: access, modify, slice")
    operation = input("Enter operation: ")

    # Step 3: Perform chosen operation
    if operation == "access":
        index = int(input("Enter index to access: "))
        print(access_element(lst, index))

    elif operation == "modify":
        index = int(input("Enter index to modify: "))
        new_value = input("Enter new value: ")
        # Convert new_value to integer if possible
        if new_value.isdigit():
            new_value = int(new_value)
        print(modify_element(lst, index, new_value))

    elif operation == "slice":
        start = int(input("Enter start index: "))
        end = int(input("Enter end index: "))
        print(slice_list(lst, start, end))

    else:
        print("Invalid operation.")

# Run the game
if __name__ == "__main__":
    index_game()
```

# *list_practice*

```python
"""
Performs list operations such as calculating length,
appending an element, and printing all elements.
"""

def main():
    # Step 1: Create the fruit list
    fruit_list = ['apple', 'banana', 'orange', 'grape', 'pineapple']

    # Step 2: Print the length of the list
    print("Length of fruit list:", len(fruit_list))

    # Step 3: Append 'mango' to the list
    fruit_list.append('mango')

    # Step 4: Print each fruit in the updated list
    print("Updated fruit list:")
    for fruit in fruit_list:
        print(fruit)

# Run the program
if __name__ == '__main__':
    main()
```
